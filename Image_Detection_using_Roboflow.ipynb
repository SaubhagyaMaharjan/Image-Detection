```python
from ultralytics import YOLO

# Load a YOLOv8 model (choose between yolov8n, yolov8s, yolov8m, yolov8l, yolov8x)
model = YOLO("yolov8m.pt")
# Train the model
model.train(data="/content/Blue-Cup-1/data.yaml",epochs=50,imgsz=640,batch=16)
```

```python
# Path to your trained model (best checkpoint)
model_path = "/content/runs/detect/train/weights/best.pt"

# Load best trained weights
model = YOLO(model_path)
```

```python
 # Evaluate performance on validation set
metrics = model.val()

# Shows mAP, precision, recall, etc.
print(metrics)
```

```python
# Inference on a new single image
image_path = "/content/Blue-Cup-1/train/images/IMG_20250612_211257_jpg.rf.ef80085b74775baa2cd40e3d9500af41.jpg" # Has Cup


results = model(image_path, conf=0.19) # Has Cup


results[0].show()

# Displays image with bounding boxes

boxes = results[0].boxes
if boxes is not None and len(boxes) > 0:
        model(
        source=image_path,
        save=True,
        save_dir="/content/runs/detect/predict"
        )
        print("Saved with predictions")
else:
    print("No predictions found!")
```
